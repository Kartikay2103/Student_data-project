## **Dataset Overview**

The analysis is based on the **Student_data** dataset, which includes the following key variables:

* **Gender**: Contains data for two genders – **male** and **female**.

* **Ethnicity**: Categorised into different ethnic groups – **Group A, Group B, Group C, Group D**, and **Group E**.

* **Parental Education Level**: Represents the education levels of parents of both male and female students.

* **Scores**: Includes student performance scores in three subjects – **Math, Reading, and Writing** – for both genders.

* **Training Institution**: Refers to the **name of the institution** where male and female students are enrolled.


## **Objectives of the Analysis**

* To **analyze and compare the average scores** provided by different **training institutions**.

* To **evaluate student performance** across three subjects: **Math, Reading, and Writing**.

* To **examine the count of students** by **ethnicity**, categorized by **both gender (male and female)**.

## EDA---> Exploratory Data Analysis


```python
# Barplot
sns.barplot(x='training_institute',y='total_score',data=df_student,color='yellow')
plt.title('Avg Total Score by training Institute')
plt.ylabel('Avg Total Score')
plt.xlabel('Training Institute')
```

## Insight
The graph titled "Avg Total Score by Training Institute" displays the Training Institutes on the X-axis and the Average Total Score on the Y-axis. It shows that both institutes have comparable student outcomes in terms of total scores, with minimal variation within each group.

```python
# Boxplot
sns.boxplot(x= 'gender', y='writing_score',color='Green',data=df_student)
plt.title("Writing Score Distribution by Gender")
plt.xlabel("Writing_score")
plt.ylabel("Frequency")
plt.show()
```

## Insights
In the boxplot titled 'Writing Score Distribution by Gender', the writing_score is shown on the X-axis and Frequency on the Y-axis. The graph indicates that there is no significant difference between male and female students in writing scores, their score distributions, averages, and variances are nearly identical. Both groups have a similar spread of scores and a few low outliers.


```python
# Histogram
sns.histplot(df_student['math_score'], bins=10, kde=True, color="Red")
plt.title('Math Score Distribution')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

## Insights
In the histogram titled 'Math Score Distribution', the Math Score is shown on the X-axis and Frequency on the Y-axis. the graph tells about the maths score of both male and female students The math score distribution is roughly normal, centered around 65–75, with most students scoring in this mid-to-high range. Scores range from 20 to 100, with few extreme low or high values. Overall, performance is consistent and balanced.

```python
# Pairplot
sns.pairplot(df_student, vars=['math_score', 'reading_score', 'writing_score', 'total_score'], hue='gender', diag_kind='kde')
plt.show()
```

## Insights
The pairplot shows the distribution of math, reading, writing, and total scores for all male and female students.   
- Math, reading, and writing scores are all positively correlated.
- Total scores correlate highly with individual subject scores.
- No significant gender differences in score distributions or relationships.
- Score distributions for all subjects are roughly normal and similar between genders.
- No strong clustering or outliers by gender.



```python
# Scatterplot
sns.scatterplot(x='reading_score', y='writing_score', hue='test_prep_course', data=df_student)
plt.title('Reading vs Writing Scores by Test Prep Course')
plt.show()
```

## Insights
The Scatterplot shows the 'reading vs writing scores by test prep course', with completed and non-completed courses. The x-axis shows reading score, and the y-axis shows writing score.
* **Positive Correlation**: Higher reading scores align with higher writing scores, showing strong literacy linkage.
* **Test Prep Impact**: Students who completed the prep course generally score higher, but there's notable overlap.
* **Score Range**: Most scores fall between 50–90; very few are extremely high or low.
* **Outliers**: No major outliers observed; data is fairly consistent.


```python
# Countplot
sns.countplot(x='ethnicity', hue='gender', data=df_student)
plt.title('Count of Students by Ethnicity and Gender')
plt.show()
```

## Insights
The countplot shows the 'Count of students by ethnicity and gender', males and females. The x-axis shows ethnicity, and the y-axis shows count.
**Group C Dominates**: Group C has the highest student count for both females and males, with females slightly leading.
**Gender Imbalance by Group**: Females outnumber males in Groups B and C.Males outnumber females in Groups A, D, and E (though only slightly in D and E).
**Group A is Smallest**: Group A has the lowest number of students, especially among females.
**Balanced Distribution**: Group D shows a nearly equal gender split, suggesting diversity.
